```python
#Plotting Bar charts on Yearly Sales
yearly_sales = yearly_sales_quantity.groupby('Invoice_Year').sum(numeric_only=True).reset_index()
sns.catplot(y='Sales Amount',x='Invoice_Year', data = yearly_sales, palette= 'bright', kind = 'bar')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.title('Yearly Sales')
yearly_sales[['Invoice_Year','Sales Amount']]
```

```python
#Plotting pie chart for Yearly sales
plt.pie('Sales Amount', labels = 'Invoice_Year', data = yearly_sales, autopct = '%1.2f%%', startangle= 90)
plt.title("Yearly Sales Percentage")
plt.legend(yearly_sales['Sales Amount'], loc = 'best')
plt.show()
```

**Observations:**  
   
1. Year 2017 has highest sales i.e.,46.98% and the Sales Amount is 87462706.4   
2. Year 2018 has least sales i.e., 11.18% and the Sales Amount is 20817471
